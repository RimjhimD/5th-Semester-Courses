Importing necessary libraries

```python
import pandas as pd
import numpy as np

```

Reading the given Netflix datas from Kaggle

```python

df = pd.read_csv('netflix_titles.csv', encoding='ISO-8859-1')

# Check the first few rows of the dataset
df.head()
```

Checking the column types.

```python
# Check the data types of each column
print(df.dtypes)
```

Checking the dataset info

```python
# Check the summary information about the dataset
df.info()
```

Finding the null values.

```python
# Find null values in each column
print(df.isnull().sum())
```

Dropping rows with NULL values

```python
# Drop rows with any null values
df_cleaned = df.dropna()

# Check if null values are dropped
print(df_cleaned.isnull().sum())
```

Dropping columns with NULL values

```python
# Drop columns with any null values
df_cleaned = df.dropna(axis=1)

# Check if any columns have been dropped
print(df_cleaned.isnull().sum())
```

After performing the 4 task,the cleaned dataset

```python
# Get summary information about the dataset
df_cleaned.info()
```
